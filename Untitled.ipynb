```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("ecommerce_customer_data_5000.csv")
df.head(5)
```

```python
df.describe()
```

```python
df.info()
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = LabelEncoder().fit_transform(df[col])
```

```python
for col in ["CustomerID", "review_text"]:
    if col in df.columns:
        df = df.drop(columns=[col])
```

```python
# STEP 4: Feature Engineering
df["customer_value"] = df["total_spent"] / (df["total_orders"] + 1)
df["engagement_score"] = df["avg_session_time"] + df["pages_per_session"] + df["email_click_rate"] * 10
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=["churn", "purchase_probability"])
y_churn = df["churn"]
```

```python
X = X.select_dtypes(include=[np.number])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_churn, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
churn_model = LogisticRegression(max_iter=5000)  
churn_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred = churn_model.predict(X_test)

```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
y_purchase = df["purchase_probability"]
```

```python
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X, y_purchase, test_size=0.2, random_state=42)
```

```python
X_train_p = scaler.fit_transform(X_train_p)
X_test_p = scaler.transform(X_test_p)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
purchase_model = RandomForestRegressor(n_estimators=100, random_state=42)
purchase_model.fit(X_train_p, y_train_p)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred_p = purchase_model.predict(X_test_p)
```

```python
print("MSE:", mean_squared_error(y_test_p, y_pred_p))
print("R2 Score:", r2_score(y_test_p, y_pred_p))
```

```python
grid = GridSearchCV(
    LogisticRegression(max_iter=5000),
    param_grid,
    cv=5,
    scoring="accuracy"
)
```
